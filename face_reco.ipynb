```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory('DatsSetFaceReco/training',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('DatsSetFaceReco/testing',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPool2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
CNN = Sequential()

CNN.add(Convolution2D(32, kernel_size=(5, 5), strides=(1, 1), input_shape=(64,64,3), activation='relu'))
CNN.add(MaxPool2D(pool_size=(2,2)))
CNN.add(Convolution2D(64, kernel_size=(5, 5), strides=(1, 1), activation='relu'))
CNN.add(MaxPool2D(pool_size=(2,2)))
CNN.add(Convolution2D(128, kernel_size=(5, 5), strides=(1, 1), activation='relu'))
CNN.add(MaxPool2D(pool_size=(2,2)))
CNN.add(Flatten())
CNN.add(Dense(64, activation='relu'))
CNN.add(Dense(7, activation='softmax'))
```

```python

CNN.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics=["accuracy"])
```

```python
CNN.fit(x = training_set, validation_data = test_set, epochs = 400)
```

```python
from keras.preprocessing import image
import numpy as np
test_image = image.load_img('DatsSetFaceReco\testing\6\IMG20221115084225.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
print(CNN.predict(test_image))
```
